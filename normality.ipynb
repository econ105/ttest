```python
import numpy as np
from scipy.stats import shapiro, anderson

# Generate some example data
data = np.random.normal(loc=0, scale=1, size=100)

# Shapiro-Wilk test
shapiro_stat, shapiro_p_value = shapiro(data)
print("Shapiro-Wilk Test:")
print("Test Statistic:", shapiro_stat)
print("p-value:", shapiro_p_value)

```

```python
import numpy as np
from scipy.stats import norm

# Parameters
mu = 28 #'loc' (location parameter): It represents the mean or average of the normal distribution. The location of the peak
sigma = 5 #'sigma'SD of the normal distribution. We set 5 here
lower_bound = 20
upper_bound = 35

# Calculate the area under the curve
area = norm.cdf(upper_bound, loc=mu, scale=sigma) - norm.cdf(lower_bound, loc=mu, scale=sigma) # scale:SD of the normal distribution.
print("Area under the curve:", area)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the mean and standard deviation of the normal distribution
mean = 0
std_dev = 1

# Generate an array of x values
x = np.linspace(-4, 4, 100) #100 is the number of equally spaced points you want to generate within the range from -4 to 4.


# Compute the corresponding y values from the normal distribution, np.sqrt is the square root function provided by numpy, np.exp is the exponential function, and np.linspace is the function used to generate an array of equally spaced values
y = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)

# Create the plot
plt.plot(x, y)

# Set the plot title and labels
plt.title('Normal Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')

# Display the plot
plt.show()
```

```python
from scipy.stats import norm

# Set the mean and standard deviation
mean = 28
std_dev = 5

# Calculate the value corresponding to the 90th percentile
percentile_value = norm.ppf(0.9, loc=mean, scale=std_dev)

# Print the result
print("Value corresponding to the 90th percentile:", percentile_value)
```

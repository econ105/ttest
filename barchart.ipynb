```python

import matplotlib.pyplot as plt

# Dataset creation.
economic_courses = ['ECON346', 'ECON105', 'FIN445', 'FIN4456', 'AI', "ECON100", "ECON346"]
students_frequency = [23, 17, 35, 29, 12, 5, 38]

# Bar graph creation.
fig, ax = plt.subplots(figsize=(10, 5))
plt.bar(economic_courses, students_frequency)
plt.title("The number of students in each courses")
plt.xlabel("Economic courses")
plt.ylabel("Frequency")
plt.show()
```

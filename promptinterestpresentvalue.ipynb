```python
def calculate_simple_interest(principal, rate, time):
    interest = (principal * rate * time) / 100
    return interest

# Example usage:
principal = 1000
rate = 5
time = 2

simple_interest = calculate_simple_interest(principal, rate, time)
print("Simple Interest:", simple_interest)
```

```python
def calculate_compound_interest(principal, rate, time):
    amount = principal * (1 + rate / 100) ** time
    interest = amount - principal
    return interest

# Example usage:
principal = 1000
rate = 5
time = 2

compound_interest = calculate_compound_interest(principal, rate, time)
print("Compound Interest:", compound_interest)
```

```python
def calculate_present_value(future_value, rate, time):
    present_value = future_value / ((1 + rate / 100) ** time)
    return present_value

# Example usage:
future_value = 1100
rate = 5
time = 2

present_value = calculate_present_value(future_value, rate, time)
print("Present Value:", present_value)
```
